```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers

from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
from keras.utils import plot_model
```

```python
df_fighters = pd.read_csv("fighters_clean.csv")
df_matches = pd.read_csv("matches_clean.csv")
```

```python
df_fighters.set_index("Name", inplace=True)
df_fighters.head()
```

```python
df_fighters['Total Fights'] = df_fighters['Wins'] + df_fighters['Losses'] + df_fighters['Draws']
df_fighters
```

```python
df_fighters.drop("Weight Class", axis=1, inplace=True)
df_fighters.drop("Wins", axis=1, inplace=True)
df_fighters.drop("Losses", axis=1, inplace=True)
df_fighters.drop("Draws", axis=1, inplace=True)
```

```python
df_fighters.head(1)
```

```python
stats = df_fighters.columns.to_list()
stats
```

```python
df_matches.head(4)
```

```python
df_matches.drop("Weight Class", axis=1, inplace=True)
```

```python
df_matches["result"] = (df_matches["Winner"] == df_matches["Red"]).astype("int")
df_matches.drop("Winner", axis=1, inplace=True)
```

```python
df_matches.head()

# if df_matches.loc[df_matches['result']==1:
#     for i =10 to 570
#         df_red = 
    
# idx = (df_matches['result'] == 1)
```

```python
# make sure all fighters are in the figher database and remove any that aren't
fighters_list = df_fighters.index.values.tolist()

df_matches = df_matches.loc[(df_matches["Red"].isin(fighters_list)) &
                                  (df_matches["Blue"].isin(fighters_list))].copy()

df_fighters.head()
```

```python
# turn df_matches into fighter 1 - fighter 2 (the differences) for training the model
for col in stats:
    df_matches[col] = df_matches.apply(
        lambda row: df_fighters.loc[row["Red"], col] - df_fighters.loc[row["Blue"], col], axis=1)
    
# drop fighter1 and fighter2 columns
df_matches.drop(["Red", "Blue"], axis=1, inplace=True)
```

```python
df_matches.head()
```

```python
print(len(df_matches.loc[df_matches["result"]==1]))
print(len(df_matches.loc[df_matches["result"]==0]))
```

```python
df_matches.to_csv(r'training_model.csv', index=False)
```

```python
X, y = df_matches.iloc[:, 1:], df_matches.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = Sequential()

model.add(Dense(16, input_dim=X_train_scaled.shape[1],
                activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(16, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['binary_accuracy'])

model.fit(x=X_train_scaled, y=y_train, epochs=200, batch_size=64, verbose=0)
test_results = model.evaluate(x = X_test_scaled, y = y_test, verbose=0)
print("Test Accuracy = {}".format(test_results[1]))
```

```python
def predict(model, Red, Blue):
    data1 = df_fighters.loc[Red, stats]
    data2 = df_fighters.loc[Blue, stats]
    data_diff = (data1 - data2).values.reshape(1, -1)
    data_diff_scaled = scaler.transform(data_diff)
    return model.predict(data_diff_scaled)

print(predict(model, "Khabib Nurmagomedov", "Tony Ferguson"))
```

```python
fight1 = predict(model,"Derrick Lewis", "Blagoy Ivanov")
fight2 = predict(model,"Blagoy Ivanov", "Derrick Lewis")
print(fight1)
print(fight2)
```
